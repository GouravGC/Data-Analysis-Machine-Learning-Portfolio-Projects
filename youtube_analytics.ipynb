```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
import pandas as pd
import random
from faker import Faker
from datetime import datetime, timedelta

fake = Faker()

categories = ['Education', 'Tech', 'Gaming', 'Lifestyle', 'Vlogs', 'Entertainment']
tags_pool = ['python', 'tutorial', 'react', 'study', 'coding', 'review', 'travel', 'productivity', 'javascript', 'motivation', 'funny', 'vlog']

data = []
for i in range(150):
    upload_date = fake.date_time_between(start_date='-90d', end_date='now')
    views = random.randint(500, 50000)
    likes = int(views * random.uniform(0.03, 0.1))
    dislikes = int(likes * random.uniform(0.01, 0.1))
    comments = int(views * random.uniform(0.01, 0.05))
    duration = random.randint(180, 3600)  # 3 min to 1 hour
    tags = ','.join(random.sample(tags_pool, k=random.randint(2, 5)))
    category = random.choice(categories)
    subscribers_gained = random.randint(1, 100)

    data.append([
        f"VID{i+1:03d}", fake.sentence(nb_words=4), upload_date, duration,
        views, likes, dislikes, comments, tags, category, subscribers_gained
    ])

df = pd.DataFrame(data, columns=[
    'Video_ID', 'Title', 'Upload_Date', 'Duration',
    'Views', 'Likes', 'Dislikes', 'Comments', 'Tags',
    'Category', 'Subscribers_Gained'
])

df.to_csv("youtube_data_large.csv", index=False)
print("✅ Generated dataset with 150 rows: youtube_data_large.csv")
```
