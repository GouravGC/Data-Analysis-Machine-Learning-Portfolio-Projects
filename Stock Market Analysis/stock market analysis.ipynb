```python
import pandas as pd
import plotly.graph_objs as go
import plotly.io as pio

# Optional: Use kaleido to export images
'''pio.kaleido.scope.default_format = "png"
pio.kaleido.scope.default_width = 1200
pio.kaleido.scope.default_height = 600'''

# Load your cleaned CSV
#df = pd.read_csv("TSLA_cleaned.csv")

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Set Date as index (optional, for time-series indexing)
df.set_index('Date', inplace=False)

# Rename for ease
df.rename(columns={
    'Close': 'C',
    'High': 'H',
    'Low': 'L',
    'Open': 'O',
    'Volume': 'V'
}, inplace=True)

# ===============================
# 1. Interactive Line Plot - Close Price
# ===============================
fig_close = go.Figure()
fig_close.add_trace(go.Scatter(
    x=df['Date'], y=df['C'],
    mode='lines',
    name='Close Price',
    line=dict(color='royalblue'),
    hovertemplate='Date: %{x}<br>Close: %{y:.2f}'
))

fig_close.update_layout(
    title='TSLA Close Price Over Time',
    xaxis_title='Date',
    yaxis_title='Close Price',
    hovermode='x unified'
)

fig_close.write_html("tsla_close_price.html")
#fig_close.write_image("tsla_close_price.png")
fig_close.show()

# ===============================
# 2. 30-day Moving Average
# ===============================
df['C_30MA'] = df['C'].rolling(window=30).mean()

fig_ma = go.Figure()
fig_ma.add_trace(go.Scatter(x=df['Date'], y=df['C'], mode='lines', name='Close', line=dict(color='gray')))
fig_ma.add_trace(go.Scatter(x=df['Date'], y=df['C_30MA'], mode='lines', name='30-Day MA', line=dict(color='red')))

fig_ma.update_layout(
    title='TSLA Close Price with 30-day Moving Average',
    xaxis_title='Date',
    yaxis_title='Price',
    hovermode='x unified'
)

fig_ma.write_html("tsla_30day_ma.html")
#fig_ma.write_image("tsla_30day_ma.png")
fig_ma.show()

# ===============================
# 3. Volume Over Time
# ===============================
fig_vol = go.Figure()
fig_vol.add_trace(go.Bar(
    x=df['Date'],
    y=df['V'],
    marker=dict(color='orange'),
    name='Volume',
    hovertemplate='Date: %{x}<br>Volume: %{y:,}'
))

fig_vol.update_layout(
    title='TSLA Volume Over Time',
    xaxis_title='Date',
    yaxis_title='Volume',
    hovermode='x unified'
)

fig_vol.write_html("tsla_volume.html")
#fig_vol.write_image("tsla_volume.png")
fig_vol.show()

# ===============================
# 4. Candlestick Chart
# ===============================
fig_candle = go.Figure(data=[go.Candlestick(
    x=df['Date'],
    open=df['O'],
    high=df['H'],
    low=df['L'],
    close=df['C'],
    name='Candlestick',
    increasing_line_color='green',
    decreasing_line_color='red'
)])

fig_candle.update_layout(
    title='TSLA Candlestick Chart',
    xaxis_title='Date',
    yaxis_title='Price',
    hovermode='x unified'
)

fig_candle.write_html("tsla_candlestick.html")
#fig_candle.write_image("tsla_candlestick.png")
fig_candle.show()
```

```python
import pandas as pd
import plotly.graph_objs as go

# Load your cleaned CSV
#df = pd.read_csv("TSLA_cleaned.csv")

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Rename columns for easier use
df.rename(columns={
    'Close': 'C',
    'High': 'H',
    'Low': 'L',
    'Open': 'O',
    'Volume': 'V'
}, inplace=True)

# Optionally, reduce size for testing speed
# df = df.tail(500)

# ========== FUNCTION DEFINITIONS ==========

def plot_close_price(df):
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df['Date'], y=df['C'],
        mode='lines',
        name='Close Price',
        line=dict(color='royalblue'),
        hovertemplate='Date: %{x}<br>Close: %{y:.2f}'
    ))
    fig.update_layout(
        title='TSLA Close Price Over Time',
        xaxis_title='Date',
        yaxis_title='Close Price',
        hovermode='x unified'
    )
    fig.write_html("tsla_close_price.html")
    fig.show()


def plot_30day_moving_average(df):
    df['C_30MA'] = df['C'].rolling(window=30).mean()
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df['Date'], y=df['C'], mode='lines', name='Close', line=dict(color='gray')))
    fig.add_trace(go.Scatter(x=df['Date'], y=df['C_30MA'], mode='lines', name='30-Day MA', line=dict(color='red')))
    fig.update_layout(
        title='TSLA Close Price with 30-Day Moving Average',
        xaxis_title='Date',
        yaxis_title='Price',
        hovermode='x unified'
    )
    fig.write_html("tsla_30day_ma.html")
    fig.show()


def plot_volume(df):
    fig = go.Figure()
    fig.add_trace(go.Bar(
        x=df['Date'],
        y=df['V'],
        marker=dict(color='red'),
        name='Volume',
        hovertemplate='Date: %{x}<br>Volume: %{y:,}'
        
    ))
    fig.update_layout(
        title='TSLA Volume Over Time',
        xaxis_title='Date',
        yaxis_title='Volume',
        hovermode='x unified'
    )
    fig.write_html("tsla_volume.html")
    fig.show()


def plot_candlestick(df):
    fig = go.Figure(data=[go.Candlestick(
        x=df['Date'],
        open=df['O'],
        high=df['H'],
        low=df['L'],
        close=df['C'],
        name='Candlestick',
        increasing_line_color='green',
        decreasing_line_color='red'
    )])
    fig.update_layout(
        title='TSLA Candlestick Chart',
        xaxis_title='Date',
        yaxis_title='Price',
        hovermode='x unified'
    )
    fig.write_html("tsla_candlestick.html")
    fig.show()

# ========== CALL ONLY ONE AT A TIME ==========

plot_close_price(df)
plot_30day_moving_average(df)
plot_volume(df)
plot_candlestick(df)
```
